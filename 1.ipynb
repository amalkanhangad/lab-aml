```python
import numpy as np
import matplotlib.pyplot as plt
def gradient_decent(X, y, learning_rate, epochs):
    m=0
    b=0
    n=len(X)
    costs=[]

    
    for epoch in range(epochs):
        # predcition
        y_pred = m*X + b
        error = y_pred - y

        # update m and b
        m -= learning_rate * (1/n) *np.sum(X * error)
        b -= learning_rate * (1/n) *np.sum(error)

        #calculate cost and store it in every 100 epochs
        if epoch % 100 ==0:
            cost = (1/(2*n)) * np.sum(error **2)
            costs.append(cost)
            print(f'epoch {epoch}, Cost {cost}')

    #plot epoch vs cost
    plt.plot(range(0, epochs, 100), costs,marker='o')
    plt.xlabel('Epochs') 
    plt.ylabel('Cost') 
    plt.title('Cost vs Epochs')
    plt.show()
    return m,b

# generate some random data for demonstration
np.random.seed(42)
X = np.random.rand(100)
y = 2 + 3 * np.random.randn(100) * 0.1
learning_rate = 0.1
epochs = 1000

m,b = gradient_decent(X, y, learning_rate, epochs)

print("slope(m):",m)
print("intercept(m)",b)


```
